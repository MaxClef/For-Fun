```python
#text to speech library
import pyttsx3

def tt_sp(text):
    '''
    Docstring: text to speech
    '''
    rate = 185
    engine = pyttsx3.init()
    engine.setProperty("rate", rate)
    engine.say(text)
    return engine.runAndWait()

def action_1():
    '''
    Docstring: greet function and action response
    '''
    print("Hi, I'm a moody chatbot")
    print("I will only multiply 2 numbers for you, since I'm working for free")      
    name_1 = input("Now that you know what you'll get out of me, what's your name? ")
    first_name = name_1.split()[0].capitalize()
    last_name = None
    name_to_use = None  
          
    if len(name_1.split()) < 2:
        last_name = input("And what's your last name?").capitalize()
    else:
        last_name = name_1.split()[-1].capitalize()
    
    
    if len(first_name) == len(last_name):
        name_to_use = f"{first_name[0]}.{last_name[0]}."
        print(f"They both seem compliocated... I'll call you {name_to_use}")
    elif len(first_name) < len(last_name):
        name_to_use = first_name
        print(f"{last_name} is way to long. I'll call you {name_to_use}")
    else:
        name_to_use = last_name
        print(f"{first_name} is way to long. I'll call you {name_to_use}")
        
    answer_1 = input(f"So {name_to_use}, do you need 2 numbers nultiplied? yes/no:")
    
    if answer_1 == "yes":
        print("Oh GOSH... ok")
    elif answer_1 == "no":
        print("Great, I have better things to do")
    else:
        pass
    
    return answer_1.lower()



def action_2():
    '''
    Docstring: function to get the correct action (yes/no)
    '''
    answer = input("Please check your spelling 'yes' or 'no' !!!:")
    response = ["yes", "no"]
    numbers = list(range(10))
    n_tries = 0
    
    while answer not in response:
        if n_tries < 1:
            answer = input("COME ON... Can you PLEASE type in 'yes' or 'no': ")
        else:
            answer = input("COME ON... Can you PLEASE type in 'yes' or 'no': ")
        n_tries += 1
    return answer.lower()



def action_3():
    '''
    Docstring: function to check if the action is to be repeated
    '''
    answer = input("Do you really need to do another operation? (yes/no): ")
    
    if answer.lower() != "yes" or answer.lower() != "no":
        pass
    else:
        answer = input("Please provide 'yes' or 'no'!!!: ")
    return answer.lower()




def simple_chat_1():
    '''
    Docstring: simple chatbot that can multiply two numbers for you
    '''
    response = ["yes", "no"]
    numbers = list(range(10))
    answer = action_1()
    
    num_def = "https://en.wikipedia.org/wiki/Number"
    
    while answer not in response: 
        answer = action_2()
        
    num_tries = 1 
    num_1 = None
    num_2 = None
    
    while answer == "yes": 
        try:
            num_1 = float(input("So, what's your 1st number: "))
            num_2 = float(input("And the 2nd number: "))
        except:
            if num_tries < 2:
                print("ERORR!!! Make sure you're entering only numbers!!!")
            if num_tries == 2:
                print("REALLY... You don't know what a number is!!!!")
            elif num_tries > 2:
                print(f"READ THIS AND THAN COME BACK: {num_def}")
            num_tries += 1
        else: 
            print(f"{num_1} * {num_2} =  {num_1 * num_2}")   
            answer = action_3()
            if answer not in response:
                answer = action_2()
    else:
        txt_1 = "THANKS FOR WAISTING MY TIME. I bet you didn't know that I can speak. "
        txt_2 = "anyway. Run the cell again. if. and only if. you think you haven't waisted my time enough."
        txt_3 = None
        final_res = None
        if num_1 in numbers and num_2 in numbers:
            txt_3 = f"..   Also!..  {int(num_1)}? times {int(num_2)}?....      really?"
            final_res = txt_1 + txt_2 + txt_3
        else:
            final_res = txt_1 + txt_2
        return tt_sp(final_res)
    
  
simple_chat_1()
```
